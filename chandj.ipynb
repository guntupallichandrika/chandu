Question3: Create Pipeline, remove stop words, word frequency analysis and word clouds

```python
pipeline = [str.lower, tokenize, remove_stop]
def prepare(text, pipeline):
    tokens = text
    for fun in pipeline:
        tokens = fun(tokens)
    return tokens
```

```python
df['tokens'] = df['full_name'].apply(prepare, pipeline=pipeline)
```

```python
import re

strings = [
    "john.doe@example.com",
    "+1 (555) 123-4567",
    "2023-06-10",
    "192.168.0.1",
    "1234-5678-9012-3456",
    "invalid@email",
    "1234567890",
    "2023/06/10",
    "256.0.0.1",
    "1234-5678-9012-345"
]

patterns = {
    "Email": r"^[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}$",
    "Phone": r"^\+?1?\s*[-.\s]?\(?[2-9]\d{2}\)?[-.\s]?\d{3}[-.\s]?\d{4}$",
    "Date": r"^\d{4}-\d{2}-\d{2}$",
    "IP": r"^(\d{1,3}\.){3}\d{1,3}$",
    "Credit Card": r"^(?:\d{4}[-\s]?){3}\d{4}|\d{16}$"
}

sorted_strings = {pattern: [] for pattern in patterns}

for string in strings:
    for pattern, regex in patterns.items():
        if re.match(regex, string):
            sorted_strings[pattern].append(string)
            break

for pattern, matched_strings in sorted_strings.items():
    print(f"{pattern}:")
    for string in matched_strings:
        print(f"  {string}")
```

```python
import re

# Sample strings containing the patterns
email_example = "Please contact us at support@example.com for further assistance."
date_example = "The report was published on 12/25/2021 and updated on 01/05/2022."
phone_example = "For more information, call 123-456-7890 or visit our website."
url_example = "Check out our website at https://www.example.com for more details."
ip_example = "The server with IP address 192.168.1.1 responded in 4 ms."

# Regex patterns
email_pattern = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'
date_pattern = r'\b\d{2}/\d{2}/\d{4}\b'
phone_pattern = r'\b\d{3}[-.]\d{3}[-.]\d{4}\b'
url_pattern = r'https?://(?:www\.)?\S+(?:/|\b)'
ip_pattern = r'\b(?:\d{1,3}\.){3}\d{1,3}\b'

# Extracting and printing matches
print("Email:", re.findall(email_pattern, email_example))
print("Date:", re.findall(date_pattern, date_example))
print("Phone:", re.findall(phone_pattern, phone_example))
print("URL:", re.findall(url_pattern, url_example))
print("IP:", re.findall(ip_pattern, ip_example))
```

```python
import pandas as pd
from textblob import TextBlob
import matplotlib.pyplot as plt

# Sample data (replace this with your actual dataset)
data = {
    'full_name': ['LeBron James', 'Kawhi Leonard', 'Giannis Antetokounmpo', 'Kevin Durant', 'James Harden']
}

df = pd.DataFrame(data)

# Function to perform sentiment analysis using TextBlob
def analyze_sentiment(text):
    blob = TextBlob(text)
    sentiment_score = blob.sentiment.polarity
    if sentiment_score > 0:
        return 'Positive'
    elif sentiment_score < 0:
        return 'Negative'
    else:
        return 'Neutral'

# Apply sentiment analysis to the 'full_name' column
df['sentiment'] = df['full_name'].apply(analyze_sentiment)

# Filter the DataFrame to include only positive, negative, and neutral sentiments
filtered_df = df[df['sentiment'].isin(['Positive', 'Negative', 'Neutral'])]

# Count the number of occurrences of each sentiment
sentiment_counts = filtered_df['sentiment'].value_counts()

# Plot the distribution of sentiments
plt.figure(figsize=(8, 6))
sentiment_counts.plot(kind='bar', color=['green', 'red', 'blue'])
plt.title('Distribution of Sentiments')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```
